```python
import pandas as pd
import numpy as np

data_location = '../../../../../properatti.csv'

data = pd.read_csv(data_location)
data.head()
```

```python
# data que elegimos para laburar

df_nuevo1 = data.loc[:,['property_type', 'country_name', 'state_name', 'price', 'currency', 'price_aprox_usd', 'surface_total_in_m2','surface_covered_in_m2','price_per_m2', 'rooms','description', 'title']]
df_nuevo1.head()
```

```python
# dataframes de superficie

df_surface = data.loc[:,['surface_total_in_m2','surface_covered_in_m2']]
df_surface.head()
```

```python
# porcentaje de nulos por columna

df_surface.isnull().sum()/df_surface.shape[0]
```

```python
# cambiamos los NaN de 'surface_total_in_m2' por ceros en otra columna

df_surface['total_zeros'] = df_surface['surface_total_in_m2'].fillna(0)
df_surface.head()
```

```python
# cambiamos los NaN de 'surface_covered_in_m2' por ceros en otra columna


df_surface['covered_zeros'] = df_surface['surface_covered_in_m2'].fillna(0)
df_surface.head()
```

```python
# máscara para laburar con los ceros de columna total

mask_total_zeros = df_surface['total_zeros']==0
total_zeros_1 = df_surface[mask_total_zeros]
total_zeros_1
```

```python
# máscara para laburar con los ceros de columna covered


mask_covered_zeros = df_surface['covered_zeros']==0
covered_zeros_1 = df_surface[mask_covered_zeros]
covered_zeros_1
```

```python
# no me acuerdo para qué hice esta creo que al pedo

df_compare = df_surface.apply(lambda x : 0 if x['total_zeros']==0 and x['covered_zeros']==0 else True, axis=1)
df_surface['zeros'] = df_compare
df_surface.head()
```

```python
# cantidad de 'total_zeros' sin data pero con data en 'covered'

df_fill_total = df_surface.apply(lambda x : 0 if x['total_zeros']==0 and x['covered_zeros'] != 0 else 1, axis=1)
df_fill_total.value_counts()
```

```python
# cantidad de 'covered' sin data pero con data en 'total_zeros'


df_fill_covered = df_surface.apply(lambda x : 0 if x['covered_zeros']==0 and x['total_zeros'] != 0 else 1, axis=1)
df_fill_covered.value_counts()
```

```python
# máscara de valores no nulos de total_zeros para sacar promedio de función a aplicar | RESUELTAS EN CÓDIGO ABAJO

total_values_mask = df_surface['total_zeros']!= 0
total_values = df_surface[total_values_mask]
total_values.head()
```

```python
# máscara de valores no nulos de covered_zeros para sacar promedio de función a aplicar | RESUELTAS EN CÓDIGO ABAJO

covered_values_mask = df_surface['covered_zeros']!= 0
covered_values = df_surface[total_values_mask]
covered_values.head()
```

```python
# CÓDIGO RESUELTO - relación entre total contra covered en columna nueva

relation_values = df_surface.apply(lambda x : 0 if x['covered_zeros']==0 or x['total_zeros'] == 0 else x['total_zeros']/ x['covered_zeros'], axis=1)
df_surface['relation_total_covered'] = relation_values
df_surface.head()
```

```python
# reemplazamos los ceros por NaN para sacar la mediana

relation_total_no_zeros = df_surface['relation_total_covered'].replace(['0', 0], np.nan)
relation_total_no_zeros.head()
```

```python
#sacamos la mediana

function = relation_total_no_zeros.median()
function
```

```python
# máscara para identificar los valores de total_surface igual a 0 y covered zeros con valor

total_mask = df_surface.apply(lambda x : True if x['total_zeros']==0 and x['covered_zeros'] != 0 else False, axis=1)
total_zeros_covered_values = df_surface[total_mask].replace(['0', 0], np.nan)
total_zeros_covered_values.head()
```

```python
# reemplazamos los valores aplicando la función

total_zeros_covered_values['surface_total'] = total_zeros_covered_values['total_zeros'].fillna(df_surface.covered_zeros * function)
total_zeros_covered_values
```

```python
# máscara para identificar los valores de covered_surface igual a 0 y total_surface con valor

covered_mask = df_surface.apply(lambda x : True if x['total_zeros']!=0 and x['covered_zeros'] == 0 else False, axis=1)
covered_zeros_total_values = df_surface[covered_mask].replace(['0', 0], np.nan)
covered_zeros_total_values.head()
```

```python
# reemplazamos los valores aplicando la función

covered_zeros_total_values['surface_covered'] = covered_zeros_total_values['covered_zeros'].fillna(df_surface.total_zeros / function)
covered_zeros_total_values
```

```python
#creamos nueva columna surface_total_m2

data['surface_total_m2'] = total_zeros_covered_values['surface_total']
data.head()
```

```python
#Valores NaN antes

data['surface_total_in_m2'].value_counts(dropna=False)
```

```python
# sumamos las columnas haciendo la definitiva
# Valores NaN/ceros después

data['surface_total'] = data['surface_total_m2'].fillna(0) + data['surface_total_in_m2'].fillna(0)
data['surface_total'].value_counts()
```

```python
# agregamos la columna surface_covered_m2

data['surface_covered_m2'] = covered_zeros_total_values['surface_covered']
data['surface_covered_m2'].value_counts(dropna = False)
```

```python
# valores NaN antes

data['surface_covered_in_m2'].value_counts(dropna = False)
```

```python
# sumamos las columnas haciendo la definitiva
#valores NaN/ceros después

data['surface_covered'] = data['surface_covered_m2'].fillna(0) + data['surface_covered_in_m2'].fillna(0)
data['surface_covered'].value_counts()
```
